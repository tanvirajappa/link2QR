```python
# Import necessary libraries
import qrcode
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Function to generate QR Code
def generate_qr_code(data):
    # Create a QR code from the input data
    qr = qrcode.QRCode(
        version=1,  # controls the size of the QR code (1 is the smallest)
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=10,  # the size of each individual box in the grid
        border=4  # thickness of the border
    )

    qr.add_data(data)
    qr.make(fit=True)
    # Create an image of the QR code
    img = qr.make_image(fill='black', back_color='white') # This line was moved inside the function
    return img # This line was already inside the function
```

```python
# Function to display QR Code in Colab using Matplotlib
def display_qr_in_colab(data):
    img = generate_qr_code(data)
    # Display the QR code using Matplotlib
    plt.imshow(img, cmap='gray') # Corrected indentation
    plt.axis('off')  # Hide the axis for clean display # Corrected indentation
    plt.show() # Corrected indentation
```

```python
# Demo 1: QR code sample website
url = "https://www.example.com"
display_qr_in_colab(url)
```

```python
# Demo 2: QR code sample website
url = "https://www.youtube.com"
display_qr_in_colab(url)
```

```python
# User-Interactive Menu : choose an option.
platforms = {
    "1": "https://www.google.com",
    "2": "https://www.pinterest.com",
    "3": "https://www.cmr.edu.in",
    "4": "https://www.instagram.com"
}

print("Choose a platform:")
print("1. Google")
print("2. Pinterest")
print("3. CMR Website")
print("4. Instagram")
print("5. Custom URL")

choice = input("Enter your choice: ")

if choice == "5":
    url = input("Enter your URL: ")
else:
    url = platforms.get(choice)

if url:
    display_qr_in_colab(url)
else:
    print("Invalid choice")
```
